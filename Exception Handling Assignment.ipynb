```python
## Question 1

def div(num,den):
    try:
        res=num/den
        print(res)
    except Exception as e:
        return e
    else:
        print('Executed Successfully')
```

```python
div(5,0)
```

```python
## Question 2

def pos_out(sub,verb,obj):
    pos_list=[]
    for s in sub:
        for v in verb:
            for o in obj:
                pos_list.append(s+' '+v+' '+o)
    return pos_list
```

```python
subjects=["Americans","Indians"]
verbs=["play","watch"]
objects=["Baseball","Cricket"]
```

```python
pos_out(subjects,verbs,objects)
```
